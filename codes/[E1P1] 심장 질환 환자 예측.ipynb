```python
submit = pd.read_csv('../input/2022-ml-midterm-p1/submit.csv')
test = pd.read_csv('../input/2022-ml-midterm-p1/test.csv')
train = pd.read_csv('../input/2022-ml-midterm-p1/train.csv')
```

```python
# 심장병을 가지고 있으면 1, 없으면 0으로 구분하는 분류 문제입니다. 
# train의 taeget 2 이상의 값을 모두 1로 바꿔줍니다. 
train.loc[train.target >1 ] =1
```

```python
# 전처리 과정 -> profiling 결과를 보고 chol==564, ca ==-1, thal==-1인 열을 삭제해 줍니다.
train = train.drop(train[(train.thalach == 0) | (train.chol == 564) | (train.ca == -1) | (train.thal == -1)].index)
```

```python
# trainX, trainY 데이터를 만들어 줍니다.

trainX = train.drop(['index','target','sex'], axis=1)
trainY = train['target']
testX = test.drop(['index','sex'], axis=1)
```
