```python
train = pd.read_csv('../input/2022-ml-quiz2-p3/train.csv')
test = pd.read_csv('../input/2022-ml-quiz2-p3/test.csv')
submit = pd.read_csv('../input/2022-ml-quiz2-p3/sample_submit.csv')
```

```python
trainX = train.drop(['index', 'Unnamed: 0', 'customerID', 'Churn', 'TotalCharges'], axis=1)
trainY = train['Churn']
testX = test.drop(['index', 'Unnamed: 0', 'customerID', 'TotalCharges'], axis=1)
```

```python
trainX
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

trainX['gender'] = le.fit_transform(trainX['gender'])
testX['gender'] = le.transform(testX['gender'])

trainX['Partner'] = le.fit_transform(trainX['Partner'])
testX['Partner'] = le.transform(testX['Partner'])

trainX['Dependents'] = le.fit_transform(trainX['Dependents'])
testX['Dependents'] = le.transform(testX['Dependents'])

trainX['PhoneService'] = le.fit_transform(trainX['PhoneService'])
testX['PhoneService'] = le.transform(testX['PhoneService'])

trainX['MultipleLines'] = le.fit_transform(trainX['MultipleLines'])
testX['MultipleLines'] = le.transform(testX['MultipleLines'])

trainX['InternetService'] = le.fit_transform(trainX['InternetService'])
testX['InternetService'] = le.transform(testX['InternetService'])

trainX['OnlineSecurity'] = le.fit_transform(trainX['OnlineSecurity'])
testX['OnlineSecurity'] = le.transform(testX['OnlineSecurity'])

trainX['OnlineBackup'] = le.fit_transform(trainX['OnlineBackup'])
testX['OnlineBackup'] = le.transform(testX['OnlineBackup'])

trainX['DeviceProtection'] = le.fit_transform(trainX['DeviceProtection'])
testX['DeviceProtection'] = le.transform(testX['DeviceProtection'])

trainX['TechSupport'] = le.fit_transform(trainX['TechSupport'])
testX['TechSupport'] = le.transform(testX['TechSupport'])

trainX['StreamingTV'] = le.fit_transform(trainX['StreamingTV'])
testX['StreamingTV'] = le.transform(testX['StreamingTV'])

trainX['StreamingMovies'] = le.fit_transform(trainX['StreamingMovies'])
testX['StreamingMovies'] = le.transform(testX['StreamingMovies'])

trainX['Contract'] = le.fit_transform(trainX['Contract'])
testX['Contract'] = le.transform(testX['Contract'])

trainX['PaperlessBilling'] = le.fit_transform(trainX['PaperlessBilling'])
testX['PaperlessBilling'] = le.transform(testX['PaperlessBilling'])

trainX['PaymentMethod'] = le.fit_transform(trainX['PaymentMethod'])
testX['PaymentMethod'] = le.transform(testX['PaymentMethod'])
```

```python
tmp = trainY.copy()
tmp[tmp == 'Yes'] = 1
tmp[tmp == 'No'] = 0
tmp[tmp.isnull()] = -1
trainY = tmp.astype('int')
```

```python
from sklearn.svm import SVC
model = SVC(C=100)
model.fit(trainX, trainY)
```

```python
pred = model.predict(testX)
submit['Churn'] = pred
```

```python
submit.to_csv('result4.csv', index=False)
```
